```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 5 random datasets having normal distribution values, loc=mean, scale= standard deviation size= 100 points with 2d
dataset_1 = np.random.normal(loc=10, scale=2000, size=(100,2))
dataset_2 = np.random.normal(loc=900, scale=1000, size=(100,2))
dataset_3 = np.random.normal(loc=2000, scale=700, size=(100,2))
dataset_4 = np.random.normal(loc=-2000, scale=900, size=(100,2))
dataset_5 = np.random.normal(loc=-500, scale=2000, size=(100,2))

dataset= np.concatenate((dataset_1,dataset_2,dataset_3,dataset_4,dataset_5))
```

```python
import matplotlib.pyplot as plt
x=dataset[:,0] # datasetis a numpy array, : slices the the 2D and takes x index(i.e., 0)
y=dataset[:,1]
#plt.hist(x) #1d
plt.figure(figsize=(10, 5))
plt.scatter(x,y) #2d
plt.show()
```

```python
X=dataset  # storing dataset in x variable # different approach

plt.figure(figsize=(10, 5))
plt.scatter(
   X[:, 0], X[:, 1],
   c='green', marker='o',
   edgecolor='black', s=50 #s=size , c= color
)
plt.show()
```

```python
# Elbow (shaped) curve  
#Searching for optimal K value uisng  WCSS ( Within-Cluster Sum of Square) to check how spreadout the cluster is.
from sklearn.cluster import KMeans

wcss =[]
for i in range (2,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter =300, n_init = 10, random_state = 0) #initializes the KMean class, 
    #max_iter runs 300 times for the predefined library function which takes averages the results
    kmeans.fit(X)                 # kmean.fit is a function calling from KMean class
    wcss.append(kmeans.inertia_) #stores the intertia values (the averaged values)

# Plot the graph to visualize the Elbow Method to find the optimal number of cluster  
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
